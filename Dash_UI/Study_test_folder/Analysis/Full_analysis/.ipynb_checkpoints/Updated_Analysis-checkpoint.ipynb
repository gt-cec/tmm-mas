```python
import json
import glob
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# ==========================================
# 1. SETUP & ANSWER KEY DEFINITION
# ==========================================

# Define the answer key based on your prompt
# Note: For array answers (multiple checkboxes), we use sets for comparison.
answer_key = {
    1: {
        1: "Yes",
        2: "SW",
        3: "Robot 3 will remain in the NW quadrant and deliver its package in the NW"
    },
    2: {
        1: {"Robot 2", "Robot 3"}, # Set for order-independent comparison
        2: "Robot 3 is stuck in rough terrain",
        3: "Robot 2"
    },
    3: {
        1: "Replanning due to rough terrain",
        2: "Robot 2 will have to replan its route",
        3: "Robot 3"
    },
    4: {
        1: "NE",
        2: "Robot 3 discovered a package in the NE and is delivering it to the NW",
        3: "NW"
    },
    5: {
        1: "Just delivered a package",
        2: "Robot 2 will head towards a different package in the SW quadrant",
        3: "Robot 1 will pick it up and deliver it in the NE."
    },
    6: {
        1: {"Robot 2", "Robot 3"},
        2: "Both robots are experiencing rough terrain",
        3: "Robot 3 will return to clear the SE"
    }
}

# Helper to normalize answers for comparison (removes case sensitivity, extra whitespace)
def normalize(val):
    if isinstance(val, str):
        # Remove potential prefixes like "b. " if they exist in user data but not key
        val = val.lower().strip()
        if val.startswith("b. "): val = val[3:]
        if val.startswith("a. "): val = val[3:]
        if val.startswith("c. "): val = val[3:]
        if val.startswith("d. "): val = val[3:]
        return val
    return val

def check_answer(scenario_id, question_id, user_answer):
    correct_answer = answer_key.get(scenario_id, {}).get(question_id)
    
    if correct_answer is None:
        return False # No key found for this scenario/question
    
    # Handle List/Set answers (Multiple Selection)
    if isinstance(correct_answer, set):
        if isinstance(user_answer, list):
            # Normalize user list to set
            user_set = {normalize(x) for x in user_answer}
            # Normalize correct set
            correct_set = {normalize(x) for x in correct_answer}
            return user_set == correct_set
        return False

    # Handle String answers
    return normalize(user_answer) == normalize(correct_answer)


# ==========================================
# 2. DATA LOADING & FILTERING
# ==========================================

# List to store parsed data
tlx_data = []      # For NASA-TLX stats
accuracy_data = [] # For Right/Wrong counts

# Get all json files in current directory
files = glob.glob('*.json')
print(f"Found {len(files)} JSON files.")

valid_files_count = 0

for file_path in files:
    with open(file_path, 'r') as f:
        try:
            data = json.load(f)
        except json.JSONDecodeError:
            print(f"Skipping {file_path}: Invalid JSON")
            continue

    # 2a. Filter: Check for 'post_study' to ensure completion
    # We look for any entry in the list with type == 'post_study'
    is_complete = any(entry.get('type') == 'post_study' for entry in data)
    
    if not is_complete:
        print(f"Skipping {file_path}: Incomplete study (no post_study tag)")
        continue
        
    valid_files_count += 1
    
    # 2b. Process Entries
    for entry in data:
        etype = entry.get('type')
        
        # --- PROCESS NASA-TLX (Post Scenario) ---
        if etype == 'post_scenario':
            # Extract basic info
            framework = entry.get('framework')
            view_type = entry.get('view_type')
            scenario = entry.get('scenario')
            run_number = entry.get('run_number')
            
            # Parse TLX scores (Extract number from string "3 - Moderate")
            answers = entry.get('answers', {})
            row = {
                'file': file_path,
                'framework': framework,
                'view_type': view_type,
                'scenario': scenario,
                'run_number': run_number
            }
            
            for metric, value_str in answers.items():
                if value_str and isinstance(value_str, str):
                    # Take the first character and convert to int
                    try:
                        score = int(value_str.split(' - ')[0])
                        row[metric] = score
                    except:
                        row[metric] = None
            
            tlx_data.append(row)

        # --- PROCESS ACCURACY (Pause Question) ---
        elif etype == 'pause_question':
            scenario = entry.get('scenario')
            pause_point = entry.get('pause_point') # This maps to Question 1, 2, 3
            user_ans = entry.get('answer')
            framework = "Unknown" # Usually defined in post_scenario, might need lookahead/lookbehind if critical here
            # Note: The JSON structure has 'framework' in post_scenario, not pause_question. 
            # If you need to correlate accuracy with framework, we might need to map run_number to framework from the post_scenario entries.
            
            is_correct = check_answer(scenario, pause_point, user_ans)
            
            accuracy_data.append({
                'file': file_path,
                'scenario': scenario,
                'question': pause_point,
                'user_answer': user_ans,
                'is_correct': is_correct,
                'run_number': entry.get('run_number')
            })

print(f"\nProcessed {valid_files_count} valid complete files.")

# ==========================================
# 3. MERGE FRAMEWORK INFO INTO ACCURACY
# ==========================================
# Since 'pause_question' doesn't have 'framework' or 'view_type', 
# we create a mapping from (file, run_number) -> (framework, view_type) using the tlx_data
df_tlx = pd.DataFrame(tlx_data)
df_acc = pd.DataFrame(accuracy_data)

if not df_tlx.empty and not df_acc.empty:
    # Create mapping dictionary
    run_info_map = df_tlx.set_index(['file', 'run_number'])[['framework', 'view_type']].to_dict('index')

    def get_meta(row):
        key = (row['file'], row['run_number'])
        if key in run_info_map:
            return pd.Series(run_info_map[key])
        return pd.Series([None, None], index=['framework', 'view_type'])

    df_acc[['framework', 'view_type']] = df_acc.apply(get_meta, axis=1)

# ==========================================
# 4. ANALYSIS & VISUALIZATION
# ==========================================

# --- A. NASA-TLX STATS ---
if not df_tlx.empty:
    print("\n--- NASA-TLX Statistics (Mean Scores) ---")
    # Group by Framework and View Type
    tlx_cols = ['mental_demand', 'physical_demand', 'temporal_demand', 'performance', 'effort', 'frustration']
    
    grouped_tlx = df_tlx.groupby(['framework', 'view_type'])[tlx_cols].mean().round(2)
    display(grouped_tlx)

    # Plotting
    # Melt for seaborn
    df_melt = df_tlx.melt(id_vars=['framework', 'view_type'], value_vars=tlx_cols, var_name='Metric', value_name='Score')
    
    plt.figure(figsize=(12, 6))
    sns.barplot(data=df_melt, x='Metric', y='Score', hue='framework', ci=None, palette="viridis")
    plt.title("NASA-TLX Scores: With vs Without Framework")
    plt.ylim(0, 6) # Assuming 1-5 or 0-6 scale
    plt.show()

    plt.figure(figsize=(12, 6))
    sns.barplot(data=df_melt, x='Metric', y='Score', hue='view_type', ci=None, palette="magma")
    plt.title("NASA-TLX Scores: Map vs Text")
    plt.ylim(0, 6)
    plt.show()

# --- B. ACCURACY STATS ---
if not df_acc.empty:
    print("\n--- Accuracy Statistics ---")
    
    # Overall % Correct
    overall_acc = df_acc['is_correct'].mean() * 100
    print(f"Overall Accuracy: {overall_acc:.2f}%")
    
    # Accuracy by Scenario
    acc_by_scenario = df_acc.groupby('scenario')['is_correct'].mean() * 100
    print("\nAccuracy by Scenario (%):")
    print(acc_by_scenario)
    
    # Accuracy by Framework/View Type
    if 'framework' in df_acc.columns:
        print("\nAccuracy by Framework (%):")
        print(df_acc.groupby('framework')['is_correct'].mean() * 100)
        
        print("\nAccuracy by View Type (%):")
        print(df_acc.groupby('view_type')['is_correct'].mean() * 100)

    # Plot Accuracy
    plt.figure(figsize=(8, 5))
    acc_plot_data = df_acc.groupby(['framework', 'view_type'])['is_correct'].mean().reset_index()
    sns.barplot(data=acc_plot_data, x='framework', y='is_correct', hue='view_type')
    plt.title("Proportion of Correct Answers")
    plt.ylabel("Accuracy (0-1)")
    plt.show()
```

```python
import json
import glob
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# ==========================================
# 1. SETUP & ANSWER KEY
# ==========================================

answer_key = {
    1: {
        1: "Yes",
        2: "SW",
        3: "Robot 3 will remain in the NW quadrant and deliver its package in the NW"
    },
    2: {
        1: {"Robot 2", "Robot 3"}, # Set for order-independent comparison
        2: "Robot 3 is stuck in rough terrain",
        3: "Robot 2"
    },
    3: {
        1: "Replanning due to rough terrain",
        2: "Robot 2 will have to replan its route",
        3: "Robot 3"
    },
    4: {
        1: "NE",
        2: "Robot 3 discovered a package in the NE and is delivering it to the NW",
        3: "NW"
    },
    5: {
        1: "Just delivered a package",
        2: "Robot 2 will head towards a different package in the SW quadrant",
        3: "Robot 1 will pick it up and deliver it in the NE."
    },
    6: {
        1: {"Robot 2", "Robot 3"},
        2: "Both robots are experiencing rough terrain",
        3: "Robot 3 will return to clear the SE"
    }
}

# Helper to normalize text for grading (removes case, extra spaces)
def normalize(val):
    if isinstance(val, str):
        val = val.lower().strip()
        # Clean common prefixes if they exist
        for prefix in ["a. ", "b. ", "c. ", "d. "]:
            if val.startswith(prefix):
                val = val[len(prefix):]
        return val
    return val

def check_answer(scenario_id, pause_point, user_answer):
    # Map Pause Point 1, 2, 3 -> Level 1, 2, 3
    # Look up correct answer
    correct_answer = answer_key.get(scenario_id, {}).get(pause_point)
    
    if correct_answer is None:
        return False # No key found
    
    # Logic for Multiple Select (Set comparison)
    if isinstance(correct_answer, set):
        if isinstance(user_answer, list):
            user_set = {normalize(x) for x in user_answer}
            correct_set = {normalize(x) for x in correct_answer}
            # Allow partial credit or strict? This is STRICT match.
            return user_set == correct_set
        return False

    # Logic for Single String
    # Handle case where user_answer might be a list of 1 item for a single choice Q
    if isinstance(user_answer, list) and len(user_answer) == 1:
        user_answer = user_answer[0]
        
    return normalize(user_answer) == normalize(correct_answer)

# ==========================================
# 2. DATA PROCESSING
# ==========================================

tlx_data = []      # NASA-TLX Rows
accuracy_data = [] # Accuracy Rows

files = glob.glob('*.json')
print(f"Found {len(files)} JSON files...")

valid_count = 0

for file_path in files:
    try:
        with open(file_path, 'r') as f:
            data = json.load(f)
    except:
        print(f"Error reading {file_path}")
        continue

    # Check for completion
    if not any(e.get('type') == 'post_study' for e in data):
        continue # Skip incomplete files
        
    valid_count += 1
    
    # 1. Build a Map of Run Number -> Condition (Framework/View)
    # We need this because 'pause_question' doesn't have the condition info, 
    # but 'post_scenario' does, and they share 'run_number'.
    run_map = {}
    for entry in data:
        if entry.get('type') == 'post_scenario':
            r_num = entry.get('run_number')
            run_map[r_num] = {
                'framework': entry.get('framework'),
                'view_type': entry.get('view_type'),
                'answers': entry.get('answers', {}) # TLX answers
            }

    # 2. Extract Data
    for entry in data:
        etype = entry.get('type')
        
        # --- ACCURACY (Pause Questions) ---
        if etype == 'pause_question':
            run_num = entry.get('run_number')
            scenario = entry.get('scenario')
            pause_pt = entry.get('pause_point') # 1, 2, or 3
            user_ans = entry.get('answer')
            
            # Get condition info from our map
            condition_info = run_map.get(run_num, {'framework': 'Unknown', 'view_type': 'Unknown'})
            
            is_correct = check_answer(scenario, pause_pt, user_ans)
            
            accuracy_data.append({
                'File': file_path,
                'Scenario': scenario,
                'Level': pause_pt, # 1=L1, 2=L2, 3=L3
                'Framework': condition_info['framework'],
                'View': condition_info['view_type'],
                'Is_Correct': 1 if is_correct else 0
            })

        # --- NASA TLX (Post Scenario) ---
        elif etype == 'post_scenario':
            run_num = entry.get('run_number')
            ans = entry.get('answers', {})
            
            row = {
                'File': file_path,
                'Framework': entry.get('framework'),
                'View': entry.get('view_type')
            }
            
            # Convert "4 - High" -> 4
            for k, v in ans.items():
                if isinstance(v, str) and ' - ' in v:
                    try:
                        row[k] = int(v.split(' - ')[0])
                    except:
                        row[k] = None
                else:
                    row[k] = None
            
            tlx_data.append(row)

print(f"Successfully processed {valid_count} files.\n")

# Create DataFrames
df_acc = pd.DataFrame(accuracy_data)
df_tlx = pd.DataFrame(tlx_data)

# ==========================================
# 3. ANALYSIS & METRICS
# ==========================================

if not df_acc.empty:
    print("="*40)
    print("      ACCURACY METRICS")
    print("="*40)

    # --- TOTAL CORRECT COUNTS ---
    print("\n--- Total Correct Answers (Count) ---")
    total_counts = df_acc.groupby('Framework')['Is_Correct'].sum()
    print(total_counts)
    
    print("\n--- Total Correct Answers (Count) by View ---")
    view_counts = df_acc.groupby('View')['Is_Correct'].sum()
    print(view_counts)

    # --- LEVEL 1, 2, 3 BREAKDOWN (With vs Without) ---
    print("\n--- Accuracy (%) by Level: With vs Without Framework ---")
    level_fw = df_acc.groupby(['Level', 'Framework'])['Is_Correct'].mean() * 100
    print(level_fw.unstack())

    # --- LEVEL 1, 2, 3 BREAKDOWN (Map vs Text) ---
    print("\n--- Accuracy (%) by Level: Map vs Text ---")
    level_view = df_acc.groupby(['Level', 'View'])['Is_Correct'].mean() * 100
    print(level_view.unstack())

if not df_tlx.empty:
    print("\n" + "="*40)
    print("      NASA-TLX METRICS (Lower is Better except Perf)")
    print("="*40)
    
    tlx_cols = ['mental_demand', 'physical_demand', 'temporal_demand', 'performance', 'effort', 'frustration']
    
    print("\n--- Mean Scores: With vs Without Framework ---")
    print(df_tlx.groupby('Framework')[tlx_cols].mean().round(2))

    print("\n--- Mean Scores: Map vs Text ---")
    print(df_tlx.groupby('View')[tlx_cols].mean().round(2))


# ==========================================
# 4. PLOTTING
# ==========================================
sns.set_style("whitegrid")

# Plot 1: Accuracy by Level (Framework comparison)
if not df_acc.empty:
    plt.figure(figsize=(10, 6))
    sns.barplot(data=df_acc, x='Level', y='Is_Correct', hue='Framework', palette='viridis', errorbar=None)
    plt.title('Accuracy by Level: Framework Comparison')
    plt.ylabel('Accuracy (0.0 - 1.0)')
    plt.xlabel('Level (Question Number)')
    plt.ylim(0, 1.1)
    plt.legend(title='Framework')
    plt.tight_layout()
    plt.show()

# Plot 2: Accuracy by Level (View comparison)
if not df_acc.empty:
    plt.figure(figsize=(10, 6))
    sns.barplot(data=df_acc, x='Level', y='Is_Correct', hue='View', palette='magma', errorbar=None)
    plt.title('Accuracy by Level: View Comparison')
    plt.ylabel('Accuracy (0.0 - 1.0)')
    plt.xlabel('Level (Question Number)')
    plt.ylim(0, 1.1)
    plt.legend(title='View Type')
    plt.tight_layout()
    plt.show()

# Plot 3: NASA-TLX Overview
if not df_tlx.empty:
    # Melt for easier plotting
    df_melt = df_tlx.melt(id_vars=['Framework', 'View'], value_vars=tlx_cols, var_name='Metric', value_name='Score')
    
    plt.figure(figsize=(12, 6))
    sns.barplot(data=df_melt, x='Metric', y='Score', hue='Framework', palette='coolwarm', errorbar=None)
    plt.title('NASA-TLX Scores: With vs Without Framework')
    plt.xticks(rotation=45)
    plt.ylim(0, 6) # Assuming 1-5 or 0-6 scale
    plt.tight_layout()
    plt.show()
```

```python
import json
import glob
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from datetime import datetime

# ==========================================
# 1. SETUP & CONFIGURATION
# ==========================================

# Answer Key for the scenarios
answer_key = {
    1: {
        1: "Yes",
        2: "SW",
        3: "Robot 3 will remain in the NW quadrant and deliver its package in the NW"
    },
    2: {
        1: {"Robot 2", "Robot 3"}, # Set for order-independent comparison
        2: "Robot 3 is stuck in rough terrain",
        3: "Robot 2"
    },
    3: {
        1: "Replanning due to rough terrain",
        2: "Robot 2 will have to replan its route",
        3: "Robot 3"
    },
    4: {
        1: "NE",
        2: "Robot 3 discovered a package in the NE and is delivering it to the NW",
        3: "NW"
    },
    5: {
        1: "Just delivered a package",
        2: "Robot 2 will head towards a different package in the SW quadrant",
        3: "Robot 1 will pick it up and deliver it in the NE."
    },
    6: {
        1: {"Robot 2", "Robot 3"},
        2: "Both robots are experiencing rough terrain",
        3: "Robot 3 will return to clear the SE"
    }
}

# Helper to normalize text (ignore case/spacing)
def normalize(val):
    if isinstance(val, str):
        val = val.lower().strip()
        # Remove common multiple choice prefixes like "a. ", "b. "
        for prefix in ["a. ", "b. ", "c. ", "d. "]:
            if val.startswith(prefix):
                val = val[len(prefix):]
        return val
    return val

# Grading Logic
def check_answer(scenario_id, pause_point, user_answer):
    correct_answer = answer_key.get(scenario_id, {}).get(pause_point)
    if correct_answer is None: 
        return False # No key found for this question
    
    # Logic for Multiple Select (Set comparison)
    if isinstance(correct_answer, set):
        if isinstance(user_answer, list):
            user_set = {normalize(x) for x in user_answer}
            correct_set = {normalize(x) for x in correct_answer}
            return user_set == correct_set
        return False

    # Logic for Single String
    if isinstance(user_answer, list) and len(user_answer) == 1:
        user_answer = user_answer[0]
        
    return normalize(user_answer) == normalize(correct_answer)

# ==========================================
# 2. DATA LOADING
# ==========================================

tlx_data = []
accuracy_data = []

# Find all JSON files in the current directory
files = glob.glob('*.json')
print(f"Found {len(files)} JSON files to process...")

for file_path in files:
    try:
        with open(file_path, 'r') as f:
            data = json.load(f)
    except Exception as e:
        print(f"Skipping {file_path}: {e}")
        continue

    # 1. Map Run Number -> Condition & Duration
    # We need to link 'pause_question' (which has no condition info) 
    # to 'post_scenario' (which has condition info) via 'run_number'.
    run_map = {}
    for entry in data:
        if entry.get('type') == 'post_scenario':
            r_num = entry.get('run_number')
            
            # Calculate Time on Task
            try:
                end_time = datetime.fromisoformat(entry.get('timestamp'))
                start_time = datetime.fromisoformat(entry.get('segment_start_time'))
                duration = (end_time - start_time).total_seconds()
            except:
                duration = None

            run_map[r_num] = {
                'framework': entry.get('framework', 'Unknown'),
                'view_type': entry.get('view_type', 'Unknown'),
                'duration': duration,
                'answers': entry.get('answers', {})
            }

    # 2. Extract Data
    for entry in data:
        etype = entry.get('type')
        
        # --- Accuracy Data (Pause Questions) ---
        if etype == 'pause_question':
            run_num = entry.get('run_number')
            scenario = entry.get('scenario')
            pause_pt = entry.get('pause_point')
            user_ans = entry.get('answer')
            
            # Get condition info
            cond = run_map.get(run_num, {'framework': 'Unknown', 'view_type': 'Unknown'})
            
            is_correct = check_answer(scenario, pause_pt, user_ans)
            
            accuracy_data.append({
                'Participant': file_path,
                'Framework': cond['framework'],
                'View': cond['view_type'],
                'Level': pause_pt,
                'Scenario': scenario,
                'Is_Correct': 1 if is_correct else 0,
                'Result_Label': 'Correct' if is_correct else 'Wrong'
            })
            
        # --- Workload Data (NASA-TLX) ---
        elif etype == 'post_scenario':
            run_num = entry.get('run_number')
            ans = entry.get('answers', {})
            row = {
                'Participant': file_path,
                'Framework': entry.get('framework'),
                'View': entry.get('view_type'),
                'Duration_Seconds': run_map.get(run_num, {}).get('duration')
            }
            
            # Convert "4 - High" -> 4
            for k, v in ans.items():
                if isinstance(v, str) and ' - ' in v:
                    try:
                        row[k] = int(v.split(' - ')[0])
                    except:
                        row[k] = None
                else:
                    row[k] = v if isinstance(v, (int, float)) else None
            
            tlx_data.append(row)

# Create DataFrames
df_acc = pd.DataFrame(accuracy_data)
df_tlx = pd.DataFrame(tlx_data)

print(f"Processed {len(df_acc)} accuracy data points and {len(df_tlx)} TLX surveys.")

# ==========================================
# 3. ANALYSIS TABLES
# ==========================================

if not df_acc.empty:
    print("\n--- ACCURACY SUMMARY ---")
    # Group by Framework and Level (1, 2, 3)
    summary = df_acc.groupby(['Framework', 'Level']).agg(
        Total=('Is_Correct', 'count'),
        Correct=('Is_Correct', 'sum')
    )
    summary['Wrong'] = summary['Total'] - summary['Correct']
    summary['Accuracy (%)'] = (summary['Correct'] / summary['Total']) * 100
    display(summary[['Correct', 'Wrong', 'Total', 'Accuracy (%)']])

if not df_tlx.empty:
    print("\n--- WORKLOAD SUMMARY (Means) ---")
    tlx_cols = ['mental_demand', 'physical_demand', 'temporal_demand', 
                'performance', 'effort', 'frustration']
    display(df_tlx.groupby('Framework')[tlx_cols].mean().round(2))

# ==========================================
# 4. PLOTTING
# ==========================================
sns.set_theme(style="whitegrid", context="paper", font_scale=1.2)

# --- PLOT 1: Accuracy by Level ---
if not df_acc.empty:
    plt.figure(figsize=(8, 5))
    sns.barplot(data=df_acc, x='Level', y='Is_Correct', hue='Framework', 
                palette='viridis', errorbar=('ci', 68), capsize=.1)
    plt.title('Prediction Accuracy by SA Level')
    plt.ylabel('Accuracy (0.0 - 1.0)')
    plt.xlabel('SA Level (1: Perception, 2: Comp, 3: Proj)')
    plt.ylim(0, 1.1)
    plt.legend(title='Condition', loc='lower right')
    plt.tight_layout()
    plt.show()

# --- PLOT 2: NASA-TLX Profile ---
if not df_tlx.empty:
    # Melt for plotting
    df_melt = df_tlx.melt(id_vars=['Framework'], value_vars=tlx_cols, 
                          var_name='Metric', value_name='Score')

    plt.figure(figsize=(10, 6))
    sns.pointplot(data=df_melt, x='Metric', y='Score', hue='Framework', 
                  dodge=True, markers=['o', 's'], linestyles=['-', '--'], palette='deep')
    plt.title('Subjective Workload Profile (NASA-TLX)')
    plt.ylabel('Score (Low - High)')
    plt.xticks(rotation=30)
    plt.ylim(0, 6) # Assuming 0-5 or 1-5 scale
    plt.grid(True, axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

# --- PLOT 3: Time on Task ---
if not df_tlx.empty and 'Duration_Seconds' in df_tlx.columns:
    plt.figure(figsize=(6, 5))
    sns.boxplot(data=df_tlx, x='Framework', y='Duration_Seconds', palette='pastel')
    sns.stripplot(data=df_tlx, x='Framework', y='Duration_Seconds', color='black', alpha=0.5)
    plt.title('Scenario Completion Time')
    plt.ylabel('Time (Seconds)')
    plt.tight_layout()
    plt.show()

# --- PLOT 4: Correlation Matrix ---
if not df_tlx.empty:
    # Select numeric columns only
    numeric_cols = tlx_cols + ['Duration_Seconds']
    numeric_df = df_tlx[numeric_cols].dropna()
    
    if not numeric_df.empty:
        plt.figure(figsize=(8, 6))
        corr = numeric_df.corr()
        sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1, fmt=".2f")
        plt.title('Correlation: Workload & Efficiency')
        plt.tight_layout()
        plt.show()
```

```python
import json
import glob
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# ==========================================
# 1. SETUP & ANSWER KEY
# ==========================================

# Standardized Answer Key

answer_key = {
    1: {
        1: "Yes",
        2: "SW",
        3: "Robot 3 will remain in the NW quadrant and deliver its package in the NW"
    },
    2: {
        1: {"Robot 2", "Robot 3"}, # Set for order-independent comparison
        2: "Robot 3 is stuck in rough terrain",
        3: "Robot 2"
    },
    3: {
        1: "Replanning due to rough terrain",
        2: "Robot 2 will have to replan its route",
        3: "Robot 3"
    },
    4: {
        1: "NE",
        2: "Robot 3 discovered a package in the NE and is delivering it to the NW",
        3: "NW"
    },
    5: {
        1: "Just delivered a package",
        2: "Robot 2 will head towards a different package in the SW quadrant",
        3: "Robot 1 will pick it up and deliver it in the NE."
    },
    6: {
        1: {"Robot 2", "Robot 3"},
        2: "Both robots are experiencing rough terrain",
        3: "Robot 3 will return to clear the SE"
    }
}

def normalize(val):
    """Normalizes strings by removing case, whitespace, and common prefixes."""
    if isinstance(val, str):
        val = val.lower().strip()
        # Remove common prefixes to allow for variations like "b. Robot 3" vs "Robot 3"
        for prefix in ["a. ", "b. ", "c. ", "d. ", "robot "]:
            if val.startswith(prefix):
                val = val.replace(prefix, "")
        return val
    return val

def check_answer(scenario_id, pause_point, user_answer):
    """Checks user answer against the key with set logic and fuzzy string matching."""
    correct_val = answer_key.get(scenario_id, {}).get(pause_point)
    
    if correct_val is None:
        return False
        
    # Logic for Multiple Select (Set comparison)
    if isinstance(correct_val, set):
        if isinstance(user_answer, list):
            u_set = {normalize(x) for x in user_answer}
            c_set = {normalize(x) for x in correct_val}
            return u_set == c_set
        return False
    
    # Logic for Single String
    if isinstance(user_answer, list) and len(user_answer) == 1:
        user_answer = user_answer[0]
        
    u_norm = normalize(user_answer)
    c_norm = normalize(correct_val)
    
    # Fuzzy match: If the correct answer is long, allow it to be 'contained' in the user answer
    # This handles "Robot 3 is stuck in rough terrain" vs "Robot 3 is encountering bad weather" 
    # if they share keywords, or just direct comparison.
    if len(c_norm) > 10: 
        # Check for key phrases if direct match fails (optional heuristic)
        return c_norm == u_norm
        
    return u_norm == c_norm

# ==========================================
# 2. DATA LOADING & PROCESSING
# ==========================================

sagat_data = []
tlx_data = []

files = glob.glob('*.json')
print(f"Found {len(files)} JSON files. Processing...")

for file_path in files:
    try:
        with open(file_path, 'r') as f:
            data = json.load(f)
    except Exception as e:
        print(f"Skipping {file_path}: {e}")
        continue
        
    # A. Build Condition Map (Run -> Framework/View)
    # We need this because 'pause_question' entries don't have the 'framework' tag,
    # but 'post_scenario' entries do, and they share the 'run_number'.
    run_map = {}
    for entry in data:
        if entry.get('type') == 'post_scenario':
            r_num = entry.get('run_number')
            run_map[r_num] = {
                'framework': entry.get('framework'),
                'view_type': entry.get('view_type')
            }
            
            # Extract NASA-TLX while we are here
            ans = entry.get('answers', {})
            row = {
                'File': file_path,
                'Framework': entry.get('framework'),
                'View': entry.get('view_type'),
                'Scenario': entry.get('scenario')
            }
            # Parse "4 - High" to integer 4
            for k, v in ans.items():
                if isinstance(v, str) and ' - ' in v:
                    try:
                        row[k] = int(v.split(' - ')[0])
                    except:
                        row[k] = None
            tlx_data.append(row)

    # B. Extract Accuracy Data (SAGAT)
    for entry in data:
        if entry.get('type') == 'pause_question':
            run_num = entry.get('run_number')
            scenario = entry.get('scenario')
            pause_pt = entry.get('pause_point')
            user_ans = entry.get('answer')
            
            # Look up condition
            cond = run_map.get(run_num, {'framework': 'Unknown', 'view_type': 'Unknown'})
            
            # Grade Answer
            is_correct = check_answer(scenario, pause_pt, user_ans)
            
            sagat_data.append({
                'File': file_path,
                'Scenario': scenario,
                'Level': pause_pt, # Maps to SAGAT Level 1, 2, 3
                'Framework': cond['framework'],
                'View': cond['view_type'],
                'Is_Correct': is_correct,
                'User_Answer': user_ans
            })

# Create DataFrames
df_sagat = pd.DataFrame(sagat_data)
df_tlx = pd.DataFrame(tlx_data)

print("Data processing complete.")

# ==========================================
# 3. ANALYSIS: SAGAT EVALUATION
# ==========================================

if not df_sagat.empty:
    print("\n" + "="*40)
    print("      SAGAT EVALUATION RESULTS")
    print("="*40)
    
    # 1. Count Correct vs Incorrect by Level
    # This directly answers your colleague's request
    print("\n--- Raw Counts: How many questions were Wrong vs Correct? ---")
    level_counts = df_sagat.groupby(['Level', 'Is_Correct']).size().unstack(fill_value=0)
    
    # Rename columns for clarity (False=Wrong, True=Correct)
    level_counts.columns = ['Wrong', 'Correct'] if len(level_counts.columns) == 2 else level_counts.columns
    
    # Calculate Total and %
    level_counts['Total'] = level_counts.sum(axis=1)
    level_counts['Error_Rate_%'] = (level_counts['Wrong'] / level_counts['Total'] * 100).round(1)
    level_counts['Accuracy_%'] = (level_counts['Correct'] / level_counts['Total'] * 100).round(1)
    
    display(level_counts)
    
    print("\n--- Accuracy by Framework Condition ---")
    print(df_sagat.groupby('Framework')['Is_Correct'].mean().round(2))

# ==========================================
# 4. VISUALIZATIONS FOR PUBLICATION
# ==========================================
sns.set_style("whitegrid")

# Plot 1: SAGAT Accuracy by Level (The "HRI Result" Chart)
if not df_sagat.empty:
    plt.figure(figsize=(10, 6))
    # We use 'Level' on X-axis (1=Perception, 2=Comprehension, 3=Projection)
    sns.barplot(data=df_sagat, x='Level', y='Is_Correct', hue='Framework', palette='viridis', errorbar=None)
    
    plt.title('Situation Awareness (SAGAT) Accuracy by Level', fontsize=14)
    plt.ylabel('Accuracy Proportion (0-1)', fontsize=12)
    plt.xlabel('SAGAT Level (1=Perception, 2=Comprehension, 3=Projection)', fontsize=12)
    plt.axhline(0.5, color='gray', linestyle='--', alpha=0.5) # Chance line (approx)
    plt.legend(title='Condition')
    plt.tight_layout()
    plt.show()

# Plot 2: NASA-TLX Workload Profile
if not df_tlx.empty:
    tlx_cols = ['mental_demand', 'physical_demand', 'temporal_demand', 'performance', 'effort', 'frustration']
    
    # Reshape for Seaborn
    df_melt = df_tlx.melt(id_vars=['Framework', 'View'], value_vars=tlx_cols, var_name='Metric', value_name='Score')
    
    plt.figure(figsize=(12, 6))
    sns.barplot(data=df_melt, x='Metric', y='Score', hue='Framework', palette='magma', errorbar=None)
    
    plt.title('NASA-TLX Workload Profile', fontsize=14)
    plt.ylabel('Score (Lower is Better)', fontsize=12)
    plt.xlabel('Workload Component', fontsize=12)
    plt.ylim(0, 6) # Adjust based on your scale (usually 0-20 or 1-7)
    plt.legend(loc='upper right')
    plt.tight_layout()
    plt.show()
```

```python
import json
import glob
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.stats.anova import AnovaRM
from scipy import stats
import warnings

# Suppress warnings for cleaner output
warnings.filterwarnings('ignore')

# Configuration: Path to your JSON files
# Ensure this matches where your 25 files are located. 
# "*.json" looks for files in the current folder.
FILE_PATTERN = "*.json"


def load_data(file_pattern):
    files = glob.glob(file_pattern)
    records = []
    
    print(f"Found {len(files)} files. Processing...")
    
    for file in files:
        # Use filename as unique Participant ID
        pid = file 
        try:
            with open(file, 'r') as f:
                data = json.load(f)
                
            for item in data:
                # We only want the post-scenario survey data
                if item.get('type') == 'post_scenario':
                    answers = item.get('answers', {})
                    
                    # Parse "3 - Moderate" into just integer 3
                    parsed_scores = {}
                    for k, v in answers.items():
                        try:
                            # Split by space and take the first part
                            parsed_scores[k] = int(str(v).split(' ')[0])
                        except:
                            parsed_scores[k] = None
                            
                    rec = {
                        'Participant': pid,
                        'Condition_ID': item.get('condition'),
                        'View_Type': item.get('view_type'),
                        'Framework': item.get('framework'),
                        **parsed_scores
                    }
                    records.append(rec)
        except Exception as e:
            print(f"Error reading {file}: {e}")
            
    return pd.DataFrame(records)

# Execute loading
df = load_data(FILE_PATTERN)

# Check the first few rows to ensure it loaded correctly
print(f"\nDataframe Shape: {df.shape}")
print("Columns:", df.columns.tolist())
df.head()



# List of NASA-TLX subscales
subscales = ['mental_demand', 'physical_demand', 'temporal_demand', 'performance', 'effort', 'frustration']

# Dictionary to store results if you want to export them later
analysis_results = {}

for subscale in subscales:
    print(f"\n{'='*20} Analysis for: {subscale.upper()} {'='*20}")
    
    # --- 1. Repeated Measures ANOVA ---
    # Factors: View_Type * Framework
    # Subject: Participant
    try:
        aov = AnovaRM(data=df, depvar=subscale, subject='Participant', within=['View_Type', 'Framework'])
        res = aov.fit()
        print("\n--- Repeated Measures ANOVA Table ---")
        print(res.summary())
    except Exception as e:
        print(f"ANOVA could not run (check for missing data): {e}")

    # --- 2. Main Effects T-Tests (Paired) ---
    # The user specifically asked for "t values".
    # We calculate paired t-tests for the main effects by averaging across the other factor.
    
    try:
        # A. Effect of VIEW TYPE (Map vs Text)
        # Average across framework conditions for each person & view type
        view_means = df.groupby(['Participant', 'View_Type'])[subscale].mean().reset_index()
        
        # Pivot to align participants for paired test
        view_pivot = view_means.pivot(index='Participant', columns='View_Type', values=subscale).dropna()
        
        if 'map' in view_pivot.columns and 'text' in view_pivot.columns:
            t_view, p_view = stats.ttest_rel(view_pivot['map'], view_pivot['text'])
            print(f"\n> Paired T-test (Map vs Text):")
            print(f"  t = {t_view:.3f}, p = {p_view:.3f}")
            print(f"  Means: Map={view_pivot['map'].mean():.2f}, Text={view_pivot['text'].mean():.2f}")
        
        # B. Effect of FRAMEWORK (With vs Without)
        fw_means = df.groupby(['Participant', 'Framework'])[subscale].mean().reset_index()
        fw_pivot = fw_means.pivot(index='Participant', columns='Framework', values=subscale).dropna()
        
        if 'with_framework' in fw_pivot.columns and 'without_framework' in fw_pivot.columns:
            t_fw, p_fw = stats.ttest_rel(fw_pivot['with_framework'], fw_pivot['without_framework'])
            print(f"\n> Paired T-test (With vs Without Framework):")
            print(f"  t = {t_fw:.3f}, p = {p_fw:.3f}")
            print(f"  Means: With={fw_pivot['with_framework'].mean():.2f}, Without={fw_pivot['without_framework'].mean():.2f}")

    except Exception as e:
        print(f"T-tests failed: {e}")
```

```python
import json
import glob
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.stats.anova import AnovaRM
from scipy import stats
from datetime import datetime
import warnings

# Suppress warnings for cleaner output
warnings.filterwarnings('ignore')

# ==========================================
# 1. SETUP & ANSWER KEY
# ==========================================
# Extracted from your previous notebook
answer_key = {
    1: {
        1: "Yes",
        2: "SW",
        3: "Robot 3 will remain in the NW quadrant and deliver its package in the NW"
    },
    2: {
        1: {"Robot 2", "Robot 3"}, # Set for order-independent comparison
        2: "Robot 3 is stuck in rough terrain",
        3: "Robot 2"
    },
    3: {
        1: "Replanning due to rough terrain",
        2: "Robot 2 will have to replan its route",
        3: "Robot 3"
    },
    4: {
        1: "NE",
        2: "Robot 3 discovered a package in the NE and is delivering it to the NW",
        3: "NW"
    },
    5: {
        1: "Just delivered a package",
        2: "Robot 2 will head towards a different package in the SW quadrant",
        3: "Robot 1 will pick it up and deliver it in the NE."
    },
    6: {
        1: {"Robot 2", "Robot 3"},
        2: "Both robots are experiencing rough terrain",
        3: "Robot 3 will return to clear the SE"
    }
}

def normalize_answer(val):
    """Normalize strings for comparison (lowercase, strip whitespace)"""
    if isinstance(val, str):
        val = val.lower().strip()
        # Remove common prefixes if they appear in data
        for prefix in ["a. ", "b. ", "c. ", "d. "]:
            if val.startswith(prefix): val = val[len(prefix):]
        return val
    return val

def check_answer(scenario_id, question_id, user_answer):
    """Compare user answer to key"""
    correct = answer_key.get(scenario_id, {}).get(question_id)
    if correct is None: return False
    
    # Handle Multiple Choice (Set/List)
    if isinstance(correct, set):
        if isinstance(user_answer, list):
            return {normalize_answer(x) for x in user_answer} == {normalize_answer(x) for x in correct}
        return False
    
    # Handle Single String
    return normalize_answer(user_answer) == normalize_answer(correct)

# ==========================================
# 2. DATA LOADING & METRIC CALCULATION
# ==========================================
def load_and_process_data(file_pattern="*.json"):
    files = glob.glob(file_pattern)
    print(f"Found {len(files)} files. Processing...")
    
    full_data = []
    
    for file in files:
        pid = file # Filename as ID
        try:
            with open(file, 'r') as f:
                raw_data = json.load(f)
            
            # Sort events by time to ensure order
            raw_data.sort(key=lambda x: x['timestamp'])
            
            # Group by (Run, Scenario) to process as one unit
            # We use a dictionary to group items
            runs = {} 
            for item in raw_data:
                key = (item.get('run_number'), item.get('scenario'))
                if key not in runs: runs[key] = []
                runs[key].append(item)
            
            # Process each Run/Scenario
            for (run_num, scen_num), items in runs.items():
                # Find the Post-Scenario survey (contains conditions)
                post_item = next((i for i in items if i['type'] == 'post_scenario'), None)
                if not post_item: continue
                
                # --- A. METRIC: DURATION (Efficiency) ---
                # Time = End of Survey - Start of First Event in this run
                t_end = datetime.fromisoformat(post_item['timestamp'])
                t_start = datetime.fromisoformat(items[0]['timestamp'])
                duration_sec = (t_end - t_start).total_seconds()
                
                # --- B. METRIC: ACCURACY (Situation Awareness) ---
                correct_count = 0
                total_qs = 0
                for item in items:
                    if item['type'] == 'pause_question':
                        total_qs += 1
                        if check_answer(scen_num, item['pause_point'], item.get('answer')):
                            correct_count += 1
                accuracy_pct = (correct_count / total_qs * 100) if total_qs > 0 else None
                
                # --- C. METRIC: NASA-TLX (Workload) ---
                answers = post_item.get('answers', {})
                tlx_scores = {}
                for k, v in answers.items():
                    try:
                        tlx_scores[k] = int(str(v).split(' ')[0])
                    except:
                        tlx_scores[k] = None
                
                # Combine all into one record
                record = {
                    'Participant': pid,
                    'Run': run_num,
                    'Scenario': scen_num,
                    'View_Type': post_item.get('view_type'),
                    'Framework': post_item.get('framework'),
                    'Duration': duration_sec,
                    'Accuracy': accuracy_pct,
                    **tlx_scores
                }
                full_data.append(record)
                
        except Exception as e:
            print(f"Error in {file}: {e}")
            
    return pd.DataFrame(full_data)

# Load the data
df = load_and_process_data("*.json")
print(f"Data Loaded: {df.shape[0]} trials across {len(df['Participant'].unique())} participants.")

# ==========================================
# 3. STATISTICAL ANALYSIS FUNCTION
# ==========================================
def run_full_analysis(df, metric, title):
    print(f"\n{'='*10} ANALYSIS: {title.upper()} {'='*10}")
    
    # 1. Repeated Measures ANOVA
    try:
        aov = AnovaRM(data=df, depvar=metric, subject='Participant', within=['View_Type', 'Framework'])
        res = aov.fit()
        print(res.summary())
    except Exception as e:
        print(f"ANOVA skipped (likely missing data): {e}")

    # 2. Paired T-Tests for Main Effects
    try:
        # View Effect
        view_wide = df.groupby(['Participant', 'View_Type'])[metric].mean().unstack()
        if 'map' in view_wide and 'text' in view_wide:
            t_v, p_v = stats.ttest_rel(view_wide['map'], view_wide['text'], nan_policy='omit')
            print(f"T-Test View (Map vs Text): t={t_v:.3f}, p={p_v:.3f}")
            print(f"  Means: Map={view_wide['map'].mean():.2f}, Text={view_wide['text'].mean():.2f}")
            
        # Framework Effect
        fw_wide = df.groupby(['Participant', 'Framework'])[metric].mean().unstack()
        if 'with_framework' in fw_wide and 'without_framework' in fw_wide:
            t_f, p_f = stats.ttest_rel(fw_wide['with_framework'], fw_wide['without_framework'], nan_policy='omit')
            print(f"T-Test Framework (With vs Without): t={t_f:.3f}, p={p_f:.3f}")
            print(f"  Means: With={fw_wide['with_framework'].mean():.2f}, Without={fw_wide['without_framework'].mean():.2f}")
            
    except Exception as e:
        print(f"T-tests failed: {e}")

    # 3. Visualization (Interaction Plot)
    plt.figure(figsize=(6, 4))
    sns.pointplot(data=df, x='View_Type', y=metric, hue='Framework', 
                  markers=['o', 's'], capsize=.1, errorbar='se', palette='viridis')
    plt.title(f"Interaction: {title}")
    plt.ylabel(title)
    plt.xlabel("View Type")
    plt.legend(title='Framework', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(True, alpha=0.3)
    plt.show()

# ==========================================
# 4. RUN ALL METRICS  
# ==========================================

# A. Objective Metrics (New!)
run_full_analysis(df, 'Duration', 'Task Duration (Seconds)') # Lower is better
run_full_analysis(df, 'Accuracy', 'Answer Accuracy (%)')     # Higher is better

# B. Subjective Metrics (NASA-TLX)
tlx_metrics = ['mental_demand', 'physical_demand', 'temporal_demand', 'performance', 'effort', 'frustration']
for m in tlx_metrics:
    run_full_analysis(df, m, m.replace('_', ' '))
```
