```python
def select_data(processed_data, robot_id, row_index=None, column_name=None):
    """
    Selects a specific row or column from processed robot data.

    :param processed_data: Dict containing processed robot data.
    :param robot_id: ID of the robot (e.g., "robot1").
    :param row_index: (Optional) Index of the row to select.
    :param column_name: (Optional) Name of the column to retrieve.
    :return: The selected row or column values.
    """
    if robot_id not in processed_data:
        print(f"❌ Robot {robot_id} not found!")
        return None

    robot_data = processed_data[robot_id]

    # If row index is specified, return the row
    if row_index is not None:
        if 0 <= row_index < len(robot_data):
            return robot_data[row_index]
        else:
            print(f"❌ Row index {row_index} out of range for {robot_id}.")
            return None

    # If column name is specified, return the column values
    if column_name is not None:
        if column_name in robot_data[0]:  # Check if the column exists
            return [row[column_name] for row in robot_data]
        else:
            print(f"❌ Column '{column_name}' not found in {robot_id}.")
            return None

    print("⚠️ Please provide either a row index or a column name.")
    return None
```

#### Takes full csv data from Harris, reads and generates HMM data points for comparison with robot states

```python
import pandas as pd
import ast
import random

def process_robot_data(input_path, output_path):
    # Load the CSV as raw text to handle "replans" correctly
    with open(input_path, "r") as file:
        lines = file.readlines()

    # Extract headers
    headers = lines[0].strip().split(",")

    # Process only the first row of replans
    first_row = lines[1].strip().split(",", maxsplit=4)  # Ensures replans is read as one piece
    first_x, first_y, first_remaining_steps = int(first_row[0]), int(first_row[1]), int(first_row[3])
    first_replans_str = first_row[4] if len(first_row) > 4 else "[]"

    # Convert the "replans" string to a list of tuples
    try:
        first_replans_list = ast.literal_eval(first_replans_str)
    except (SyntaxError, ValueError):
        first_replans_list = []  # Default to empty if parsing fails

    # Create a DataFrame with x and y columns
    replans_df = pd.DataFrame(first_replans_list, columns=["x", "y"])

    # Insert the first row's x, y at the top
    first_row_df = pd.DataFrame([[first_x, first_y]], columns=["x", "y"])
    full_replans_df = pd.concat([first_row_df, replans_df], ignore_index=True)

    # Add formatted (x, y) column
    full_replans_df["formatted_pos"] = full_replans_df.apply(lambda row: f"({row['x']}, {row['y']})", axis=1)

    # Function to generate intervals
    def generate_intervals(index):
        base = index * 10
        if random.random() < 0.7:  # 70% chance for an increment
            base += random.choice([10, 20, 30, 50])
        return base

    # Add the 'interval' column as the last column
    full_replans_df["interval"] = [generate_intervals(i) for i in range(1, len(full_replans_df) + 1)]

    # Add the 'remaining_steps' column, keeping the first row’s remaining_steps value for all rows
    full_replans_df["remaining_steps"] = first_remaining_steps

    # Select only required columns: formatted_pos, interval, and remaining_steps
    final_df = full_replans_df[["formatted_pos", "interval", "remaining_steps"]]

    # Reformat headers
    final_df.columns = final_df.iloc[0]
    final_df = final_df.drop(0).reset_index(drop=True)

    # Save to CSV without headers and index
    final_df.to_csv(output_path, index=False, header=False)

    return final_df

# Example usage
output_path = "HMM_Robot1.csv"
processed_df = process_robot_data("robot1.csv", output_path)

# Display the result
display(processed_df)
```
